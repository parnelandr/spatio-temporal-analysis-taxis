### Changes to make of the data

* ORIGIN_CALL to int
* ORIGIN_STAND to int
* TIMESTAMP to datetime
* Remove data that has missing coordinates in POLYLINE based on MISSING_DATA field

**Calculate time taken and add as column**

```python
df['POLYLINE'].head().apply(lambda x: len(ast.literal_eval(x))*15) 
#to calculate seconds taken per trip (from count the number of coordinate pairs  x 15 sec intervals)
```

```python
df['TIME_TAKEN'] = pd.to_timedelta(df['POLYLINE'].apply(lambda x: len(ast.literal_eval(x))*15), unit='s')
```

```python
df.head()
```

```python
df.info()
```

**Calculate end time and add as column**

```python
df['END_TIMESTAMP'] = df['TIMESTAMP'] + df['TIME_TAKEN']
```
